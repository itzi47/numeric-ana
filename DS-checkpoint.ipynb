# Predicting Employee Attrition - Job Change of Data Scientists




**Significance:**
- **Financial Impact:** High attrition rates become costly due to the expenses of recruiting, hiring, and training new employees. By predicting attrition, the company will be able to introduce strategies for retention and save costs.
- **Operational Continuity:** With experience and knowledge, experienced people also leave. Attrition prediction helps plan to ensure that all critical roles are always filled.
- **Employee Morale:** High employee attrition rates are a demoralizing factor to current employees and directly result in low productivity. Being able to predict employee turnover can help in addressing the underlying problems of dissatisfaction and, hence, positively influence employee morale.

**Importance of the Solution:**
- **Proactive Retention Strategies:** Specific retention strategies can be launched targeted at the needs of employees prone to attrition through proactive identification of these employees, for instance, by providing improved work conditions, individualized incentives, and career development opportunities.
- **Improved Workforce Planning:** Predicting attrition will help enhance workforce planning and succession management; hence, the organization is prepared with a talent pipeline for critical roles.
- **Improved Organizational Stability:** Attrition, when reduced, offers a more stable workforce. This is necessary for high productivity and ensures the achievement of goals set for the organization in the long term.

Apart from being valid on a practical level, understanding and solving the problem of employee attrition would have benefits at a higher level. These assist in maintaining stable work to underpin economic stability and growth, curtail stress and anxiety on workers, and give a chance for a healthier and more inclusive working environment. Firm development of the models that forecast attrition would go a long way in setting up more nutritious and more stable workplaces that benefit individual companies and society.


## Steps for Dataset Creation

### 1. Pre-processing

In this part, I will perform several manipulations on the dataset to prepare it for use in the next stages:

#### Redundancy in the Data
I checked for duplicates in the data to see if there are any duplicate rows. I found that the data is without any duplicates, which might otherwise cause biases in the machine learning model later on.

#### Missing Values
- **Gender:** For this feature, I examined the distribution and filled missing values based on the probability of being male or female. If the generated number is smaller than the probability of being female, I filled it with "Female"; otherwise, I filled it with "Male".


```python
import random

gender_val_cont = df["gender"].value_counts()
print(gender_val_cont)
gender_val_norm = df["gender"].value_counts(normalize=True)
print(gender_val_norm)

female_percent = gender_val_cont['Female'] / (gender_val_cont['Male'] + gender_val_cont['Female'])
male_percent = 1 - female_percent

for i in range(len(df)):
    if pd.isnull(df.loc[i, 'gender']):
        if random.random() <= female_percent:
            df.loc[i, 'gender'] = "Female"
        else:
            df.loc[i, 'gender'] = "Male"




```


- **Major Discipline:** For individuals with education levels "High School" or "Primary school", the `major_discipline` feature is not applicable. Hence, I filled missing values with "No Major", and for the remaining missing values, I used the mode which is "STEM".


```python
for i in range(len(df)):
    if df.loc[i, 'education_level'] == 'High School' or df.loc[i, 'education_level'] == 'Primary School':
        df.loc[i, 'major_discipline'] = 'No Major'

df['major_discipline'].fillna("STEM", inplace=True)
```


- **Experience & Enrolled University:** I filled the missing values using the mode of each feature as they had relatively few missing values.


```python
df['experience'].fillna(">20", inplace=True)
university_val_cont = df["enrolled_university"].value_counts()
df["enrolled_university"].fillna("no_enrollment", inplace=True)
```


- **Company Size & Company Type:** i filled missing values with "Unknown" due to a large number of missing values in these features to avoid creating a bias towards any particular value.


```python
df['company_size'].fillna('Unknown', inplace=True)
df['company_type'].fillna('Unknown', inplace=True)
```

 - **Education Level & Last New Job:** Similar to the `gender` feature, I filled the missing values based on the probability of occurrence for each value in the data.


```python
education_val_cont = df['education_level'].value_counts(normalize=True)
graduate_p = education_val_cont['Graduate']
masters_p = education_val_cont['Masters'] + graduate_p

for i in range(len(df)):
    if pd.isnull(df.loc[i, 'education_level']):
        r = random.random()
        if r <= graduate_p:
            df.loc[i, 'education_level'] = 'Graduate'
        elif graduate_p < r <= masters_p:
            df.loc[i, 'education_level'] = 'Masters'
        else:
            df.loc[i, 'education_level'] = 'High School'

last_new_job_cont = df['last_new_job'].value_counts()
last_new_job_norm = df['last_new_job'].value_counts(normalize=True)

one_p = last_new_job_norm["1"]
gfour_p = last_new_job_norm[">4"]
two_p = last_new_job_norm["2"]
never_p = last_new_job_norm["never"]
four_p = last_new_job_norm["4"]
three_p = last_new_job_norm["3"]

for i in range(len(df)):
    if pd.isnull(df.loc[i, 'last_new_job']):
        r = random.random()
        if r <= one_p:
            df.loc[i, 'last_new_job'] = "1"
        elif one_p < r <= one_p + gfour_p:
            df.loc[i, 'last_new_job'] = ">4"
        elif one_p + gfour_p < r <= one_p + gfour_p + two_p:
            df.loc[i, 'last_new_job'] = "2"
        elif one_p + gfour_p + two_p < r <= one_p + gfour_p + two_p + never_p:
            df.loc[i, 'last_new_job'] = "never"
        elif one_p + gfour_p + two_p + never_p < r <= one_p + gfour_p + two_p + never_p + four_p:
            df.loc[i, 'last_new_job'] = "4"
        else:
            df.loc[i, 'last_new_job'] = "3"

df.head()
```

## Feature Extraction

I want to extract the most informative features from my dataset to improve the performance of my machine learning model. I aim to create additional features based on my knowledge and research.

### Specific - Knowledge Based

I used a specific-knowledge-based approach to create new features based on the research and domain knowledge about DS training programs. Here are the new features I created:

- **Central_city:** A binary feature that indicates if the employee comes from a city with more than 1000 employees. This feature is created based on the idea that employees from more populated cities might have different job-changing behaviors due to more job opportunities.
- **Experience_index:** A binary feature that indicates if the employee has 7 or more years of experience. I believe that having 7 or more years of experience might influence an employee's decision to stay or leave the job after completing a DS training program.


```python
city_cont = df['city'].value_counts()
Central_city = []
for i in range(len(df)):
    if city_cont[df.loc[i, 'city']] > 1000:
        Central_city.append(1)
    else:
        Central_city.append(0)
df['Central_city'] = Central_city
```

```python
Experience_index = []
for i in range(len(df)):
    if df.loc[i, 'experience'] == '>20':
        Experience_index.append('1')
    elif df.loc[i, 'experience'] == '<1':
        Experience_index.append('0')
    elif int(df.loc[i, 'experience']) >= 7:
        Experience_index.append('1')
    else:
        Experience_index.append('0')

df['Experience_index'] = Experience_index
```

### Visualizing Categorical Features vs Target

To further understand the relationship between categorical features and the target variable, I visualized the distributions of these features with respect to the target variable,I was unable to reach an unequivocal conclusion using the graphs due to a high amount of observations in certain categories. However, I chose to keep all of these variables because according to the ANOVA test, it turned out that their relationship to the target variable is significant

- **Enrolled University vs Target**
- **Education Level vs Target**
- **Major Discipline vs Target**
- **Company Size vs Target**
- **Company Type vs Target**
- **Last New Job vs Target**
- **City vs Target**
- **Experience Index vs Target**


```python
from scipy import stats
def plot_categorical_vs_target_side_by_side(column):
    # Filter out NaN values
    data = df.dropna(subset=[column, 'target'])
    counts = data.groupby([column, 'target']).size().unstack(fill_value=0)

    # Plot
    ax = counts.plot(kind='bar', figsize=(10, 6))
    ax.set_title(f'{column.replace("_", " ").capitalize()} vs Target')
    ax.set_xlabel(column.replace("_", " ").capitalize())
    ax.set_ylabel('Count')
    plt.show()

plot_categorical_vs_target_side_by_side('enrolled_university')
F, p_value = stats.f_oneway(df[df.enrolled_university == 'no_enrollment'].target,
                            df[df.enrolled_university == 'Part time course'].target,
                            df[df.enrolled_university == 'Full time course'].target)
print("Enrolled university-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)

plot_categorical_vs_target_side_by_side('education_level')
F, p_value = stats.f_oneway(df[df.education_level == 'High School'].target,
                            df[df.education_level == 'Masters'].target, df[df.education_level == 'Graduate'].target,
                            df[df.education_level == 'Phd'].target, df[df.education_level == 'Primary School'].target)
print("Education level-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)
# Major Discipline vs Target
plot_categorical_vs_target_side_by_side('major_discipline')
F, p_value = stats.f_oneway(df[df.major_discipline == 'No Major'].target,
                            df[df.major_discipline == 'STEM'].target, df[df.major_discipline == 'Humanities'].target,
                            df[df.major_discipline == 'Arts'].target, df[df.major_discipline == 'Other'].target,
                            df[df.major_discipline == 'Business Degree'].target)
print("Major discipline-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)

plot_categorical_vs_target_side_by_side('company_size')

plot_categorical_vs_target_side_by_side('company_type')
F, p_value = stats.f_oneway(df[df.company_type == 'Pvt Ltd'].target,
                            df[df.company_type == 'NGO'].target, df[df.company_type == 'Public Sector'].target,
                            df[df.company_type == 'Unknown'].target, df[df.company_type == 'Funded Startup'].target,
                            df[df.company_type == 'Early Stage Startup'].target, df[df.company_type == 'Other'].target)
print("Company type-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)

plot_categorical_vs_target_side_by_side('last_new_job')
F, p_value = stats.f_oneway(df[df.last_new_job == '1'].target,
                            df[df.last_new_job == '2'].target, df[df.last_new_job == '3'].target,
                            df[df.last_new_job == '4'].target, df[df.last_new_job == '>4'].target,
                            df[df.last_new_job == 'never'].target)
print("Last new job-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)

plt.figure(figsize=(10, 6))
data = df.dropna(subset=['city', 'target'])
counts = data.groupby(['city', 'target']).size().unstack(fill_value=0)
ax = counts.plot(kind='bar', figsize=(10, 6))
ax.set_title('City vs Target')
ax.set_xlabel('City')
ax.set_ylabel('Count')
plt.xticks(rotation=90)
plt.show()
print("City-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)

plot_categorical_vs_target_side_by_side('Experience_index')
print("Experience index-")
print("F statistic of the test: ", F, "p-value from the F distribution", p_value)
```

## Model Training

i will use the following methods for training my model and evaluating its performance:
The validation method I will use is Cross Validation. There are three ways to implement this method: k-fold, leave one out and holdout. Among these three methods, I chose to use the k-fold method. In this method, K different divisions of the data are made into a training set and a validation set, so that each sample is both in the training set (k-1 times) and in the validation set (once). After the training phase, K models are created, which are weighted into one more balanced model that reduces the bias in the test phase. On the other hand, with the holdout method, the division into percentages may not be good so that it does not adequately capture the variability in the data. In addition to this, the leave one out method leaves one sample each time for validation and allocates the rest for training, which means that a huge amount of iterations will be performed (as the number of samples in my data) compared to only K iterations in the k-fold method. In light of all this, I chose to use the k-fold method over the others.
To implement the method I will choose k=10 , and divide my data set so that each time one part of the data set will be used for validation, while the other nine parts will be used for training. In addition, I calculate the value that will be obtained in each validation according to the MSE metric - the average of the squared errors. That is, I calculate the average of the squared differences between the actual value and the value that the model estimates, and thus I can estimate the correctness of the model. Finally, I will have K MSE values, which I will average to get the metric value.

### K-Fold Cross-Validation



```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np


X = df[['city_development_index', 'training_hours', 'experience']]
y = df['target']

model = LinearRegression()

kf = KFold(n_splits=10, shuffle=True, random_state=42)
mse_values = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

average_mse = np.mean(mse_values)

print(f'K-Fold Cross-Validation MSE values: {mse_values}')
print(f'Average K-Fold Cross-Validation MSE: {average_mse}')
```

# Clustering

k-medoid is an algorithm for unsupervised learning, meaning without class labels, and its goal is to perform clustering rather than prediction or classification. Therefore I removed the target variable and adapted the dataset for the learning task. Additionally, the dataset is normalized, and categorical variables were converted to dummy variables I expect to get a model with two clusters, as the original model has two classes - 0 and 1
 I will use a feature set containing all the features except CITY and Enrolee ID, similar to the dataset on which I ran the other learning algorithms I will use the Gower distance metric to measure distance since each record in the dataset contains a combination of categorical, continuous, and binary data. The Gower distance is a number between 0 (identical) and 1 (maximally different). The Gower distance is calculated as the average of the partial differences between individuals. The general form of the coefficient is as follow
s 
I ran the algorithm and chose different Ks in the range of [2,8]

I saved the results obtained for each K examined and checked the following properties:
- Homogeneity (dense and close points to the cluster).
- Separation (good separation between cluster cenis).

I will check the existence of these properties using the following metrics:
- SiIouette - I want to maximize this score, as a higher score indicates better separation and compactness within clusters.
- DavieIBouldin - I want to minimize this score, as a lower score indicates better separation and compactness within clusters.

Based on the Davies-Bouldin index, the optimal number of clusters is 2-3, and based on the Silhouette score, the optimal number of clusters is 2 Therefore, I decided that the optimal number of clusters is K=2 Given that I obtained the expected number of clusters (2), it can be assumed that each cluster represents 0 or 1 according to the labels assigned to each sample in the original dataset.


```python
import gower
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt


gower_dist = gower.gower_matrix(X_imputed)

t
k_values = range(2, 9)

silhouette_scores = []
davies_bouldin_scores = []

for k in k_values:
    kmedoids = KMedoids(n_clusters=k, metric='precomputed', random_state=42)
    kmedoids.fit(gower_dist)
    
    labels = kmedoids.labels_
    
    silhouette_avg = silhouette_score(gower_dist, labels, metric='precomputed')
    davies_bouldin_avg = davies_bouldin_score(gower_dist, labels)
    
    silhouette_scores.append(silhouette_avg)
    davies_bouldin_scores.append(davies_bouldin_avg)
    
    print(f'For k={k}, Silhouette Score: {silhouette_avg}, Davies-Bouldin Score: {davies_bouldin_avg}')


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k')


plt.subplot(1, 2, 2)
plt.plot(k_values, davies_bouldin_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Davies-Bouldin Score')
plt.title('Davies-Bouldin Score for Different k')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
test_sizes = np.arange(0.1, 0.35, 0.05)
res2 = pd.DataFrame()

for size in test_sizes:
    X_train, X_val, y_train, y_val = train_test_split(X_imputed, y, test_size=size, random_state=123)
    dt2_model = DecisionTreeClassifier(criterion='entropy', max_depth=7, random_state=123)
    dt2_model.fit(X_train, y_train)
    
    scores2 = cross_val_score(dt2_model, X_train, y_train, cv=10, scoring='roc_auc')
    res2 = pd.concat([res2, pd.DataFrame({'size': [size], 'train_acc': [accuracy_score(y_train, dt2_model.predict(X_train))], 'val_acc': [scores2.mean()]})], ignore_index=True)

print(res2)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(res2['size'], res2['train_acc'], label='Train Accuracy', marker='o')
plt.plot(res2['size'], res2['val_acc'], label='Validation ROC AUC', marker='o')
plt.xlabel('Test Size')
plt.ylabel('Score')
plt.title('Effect of Test Size on Model Performance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
features_list = np.arange(1, 28, 1)
results = pd.DataFrame()

for k in features_list:
    selector = SelectKBest(chi2, k=k)
    X_train_selected = selector.fit_transform(X_train, y_train)
    X_val_selected = selector.transform(X_val)
    
    dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=7, random_state=123)
    dt_model.fit(X_train_selected, y_train)
    
    val_selected = cross_val_score(dt_model, X_val_selected, y_val, cv=10, scoring='roc_auc')
    results = pd.concat([results, pd.DataFrame({'k': [k], 'train_score': [accuracy_score(y_train, dt_model.predict(X_train_selected))], 'val_score': [val_selected.mean()]})], ignore_index=True)


plt.figure(figsize=(10, 5))
plt.plot(results['k'], results['train_score'], label='Train Score', marker='o')
plt.plot(results['k'], results['val_score'], label='Validation ROC AUC', marker='o')
plt.xlabel('Number of Features (k)')
plt.ylabel('Score')
plt.title('Effect of Number of Features on Model Performance')
plt.legend()
plt.grid(True)
plt.show()
```
